# Problem Statement 2:
## Write a Python program to perform the following tasks

1. Create a pandas series from the below dictionary where indices are subjects:

`{'English':{'Sam':60,'Jackson':74,'Ahree':85},
'History':{'Gloria':83,'Sam':65,'Isla':78,'Aron':72,'Gray':61},
'Geography':{'Jackson':92,'Gloria':95,'Isla':82,'Aron':75,'Ahree':76},
'Mathematics':{'Sam':99,'Gloria':74,'Jackson':89,'Ahree':85,'Gray':95},
'Science':{'Sam':89,'Aron':82,'Gray':78,'Isla':93,'Ahree':87}}`

2. Convert the created series into DataFrame and replace the null values with zeroes.

3. Transpose the DataFrame and create a new column 'Average' and fill the values in it by calculating the average of all subjects.

# Task 1: Create a pandas series from the below dictionary where indices are subjects.


```python
# The dictionary creation.
dictionary = {
'English':{'Sam':60,'Jackson':74,'Ahree':85},
'History':{'Gloria':83,'Sam':65,'Isla':78,'Aron':72,'Gray':61},
'Geography':{'Jackson':92,'Gloria':95,'Isla':82,'Aron':75,'Ahree':76},
'Mathematics':{'Sam':99,'Gloria':74,'Jackson':89,'Ahree':85,'Gray':95},
'Science':{'Sam':89,'Aron':82,'Gray':78,'Isla':93,'Ahree':87}
}


print("This is a data type: ",type(dictionary))
print('\n',"Creating the dictionary:\n",dictionary)
```

```python
# Define the new order of the series.

# the indices are defined in a list which will be used again later.
indices = ['English', 'History', 'Geography', 'Mathematics', 'Science']

# Create the series with the help of the dictionary.
series = pd.Series(dictionary, index=indices)
series # the final series
```

# Task 2: Convert the created series into DataFrame and replace the null values with zeroes.


```python
# creating the dataframe using the .from_records function.
                            # using the series and...
df = pd.DataFrame.from_records(series, index = indices)
# ... the index is the list of subjects (indices) which we defined before.

print("This is a data type: ",type(df))
print('\n',"Creating the DataFrame from the Pandas Series:")
df
```

```python
# Filling the NaN values and turning them into 0
data = df.fillna(0)
data
```

# Task 3: Transpose the DataFrame and create a new column 'Average' and fill the values in it by calculating the average of all subjects.

```python
dt = data.T # transposing the data with the .T function
dt # data transposed = dt
```

```python
avg_result = df.mean(axis=0) # axis=0 means we're calculating based on the rows (student names)

# Calculating everyone's average. (mean of entire column)
avg_result
```

```python
# Creating a new column to put the caluclated average results.
dt['Average'] = avg_result
dt
```

# Problem Statement 3: Write a Python program to create a series from 1 to 1000 and select only numbers divisible by 7 and 17

```python
# creating the series with a range from 1 to 1000. index starts from 0.
thousands = pd.Series(range(1,1000))

thousands # the series is printed as shown.
```

```python
# using a list comprehension

    # iterating over the series [thousands] of numbers.
div = [number for number in thousands
            if number % 7 == 0 or number % 17 ==0]
                # defining the conditions.

print(div) # printing what we got from the program.
```

```python
# to display all the rows when printed (you can scroll to see al results)
# pd.set_option('display.max_rows', None)

# turning the result we got into a series.
series = pd.Series(div)

# printing the list that returns all the numbers divisible by 7 and 17.
print(series)
```

```python
# Alternative solution

divisibleby = pd.Series( # putting the results found in the series into a list for displaying.
    filter( # using the filter() function in Python takes in a function and a list as arguments.
        lambda x: x % 7 == 0 or x % 17 == 0, # two of the arguments.
        thousands
    )
)

print(divisibleby)  # printing the results found for numbers divisible by 7 and 17.
```

# Problem Statement 4: Sylphia has a dataset of various cereals sold in the supermarket.



Tasks to be performed:
1. Import the dataset.
2. Plot ratings of different types of manufacturers.
3. Use xticks range form 0-100.
4. Change the style of the graph to seaborn

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
# a magic function!
```

# Task 1: importing the dataset.

```python

pd.set_option('display.max_rows', None) # displaying all rows of data.

data = pd.read_csv('Downloads/cereal.csv')
data.describe()
```

```python
data # printing the dataset.
```

# Task 2: Plot ratings of different types of manufacturers

```python
# before we plot we need to sort the data, we'll use groupby.
for group,data in data.groupby('mfr'):
    # iterating and grouping the data based on manufacturers.
    plt.barh(group,data['rating'].mean())
    # plotting the bar chart where the x values are grouped by the ratings (mean).
```

# Task 3: Changing the x axis with xticks to display a range from 0-100.

```python
for group,data in data.groupby('mfr'):
    plt.barh(group,data['rating'].mean())
plt.title('Rating for each Manufacturer')
plt.xlabel('Rating')
plt.ylabel('Manufacturer')
plt.xticks(np.arange(0,110,10)) # changing the x axis to display from 0-100 in intervals of 10.
plt.show()
```

# Task 4: Change the style of the graph to seaborn

```python
# importing seaborn.
import seaborn as sns

# using the seaborn set theme. This will change the matplot chart into a seaborn graph.
sns.set_theme()
```

```python
# Plotting everything all together.

for group,data in data.groupby('mfr'):
    plt.barh(group,data['rating'].mean())
plt.title('Rating for each Manufacturer')
plt.xlabel('Rating')
plt.ylabel('Manufacturer')
plt.xticks(np.arange(0,110,10))
plt.show()
```
